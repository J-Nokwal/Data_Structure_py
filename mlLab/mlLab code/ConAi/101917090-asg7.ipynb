```python
import requests as r
import pandas as pd
import time as t
```

```python
api_key='d6b6f511e01f1acbef2339447bdc8fb3'
BaseUrl='https://api.themoviedb.org/3/discover/movie?api_key={}&with_genres=35&primary_release_date.gte=2000&sort_by=popularity.desc'.format(api_key)
movies=[]
for i in range(1,16):
    url=BaseUrl+'&page=' + str(i)
    request= r.get(url)
    request.json()
    df=pd.DataFrame(request.json())
    mv = pd.DataFrame(request.json()['results'])

    for j in range(len(mv)):
        st = str(mv['id'][j]) + ',' + str(mv['title'][j])
        movies.append([st, mv['popularity'][j], mv['id'][j]])
movies = pd.DataFrame(movies, columns = ['Comedy Movies', 'Popularity', 'IDs'])
movies = movies.sort_values(by = 'Popularity')
movies = movies.iloc[::-1] 
moviesIds = movies['IDs']
comedy_movies = pd.DataFrame(movies['Comedy Movies'])
comedy_movies.to_csv("/home/rockbot/Documents/Data_Structure_py/mlLab/mlLab code/ConAi/comedy_movies.csv")
comedy_movies
```

```python
similar_ids=[]
for i in range(len(moviesIds)):
    url = 'https://api.themoviedb.org/3/movie/' + str(moviesIds[i]) + '/similar?api_key=' + api_key
    request = r.get(url)
    similar_movies = pd.DataFrame(request.json()['results'])
    for j in range(len(similar_movies)):
        if j < 5 and [similar_movies['id'][j], moviesIds[i]] not in similar_ids:
            # sm_ids_csv.append(str(moviesIds[i]) + ',' + str(similar_movies['id'][j]))
            similar_ids.append([moviesIds[i] ,similar_movies['id'][j]])
similar_ids_pd=pd.DataFrame(similar_ids,columns=["fromMovie","similarMovie"])
```

```python
similarpd=(similar_ids_pd.groupby(["fromMovie","similarMovie"])).size().reset_index().rename(columns={0:'count'})    
similarpd
```
