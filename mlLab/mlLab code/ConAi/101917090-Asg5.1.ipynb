```python
from selenium import webdriver
```

```python
driver = webdriver.Chrome(executable_path="/home/rockbot/Documents/Data_Structure_py/mlLab/mlLab code/chromedriver")
```

```python

driver.maximize_window()
all_books_details = {}
for pageno in range(1,3):
    driver.get("https://www.amazon.in/gp/bestsellers/books/ref=zg_bs_pg_" + str(pageno) + "?ie=UTF8&pg=" + str(pageno))
    for j in range(1, 51):
        driver.find_element_by_xpath("//*[@id='zg-ordered-list']/li[" + str(j) + "]/span/div/span/a").click()
        book_name = driver.find_element_by_xpath("//*[@id='productTitle']").text
        data = driver.find_element_by_xpath("//*[@id='detailBullets_feature_div']").text.split('\n')
        book_details = ['Publisher', 'Language', 'Paperback', 'ISBN-10', 'ISBN-13', 'Weight', 'Dimensions']
        detail_dict = dict.fromkeys(book_details)
        for key in detail_dict.keys():
            for i in range(len(data)):
                if key in data[i]:
                    detail_dict[key] = data[i].split(':')[1]
                    break
                elif key == 'Paperback':
                    if 'Hardcover' in data[i]:
                        detail_dict[key] = data[i].split(':')[1]
            else:
                detail_dict[key] = None
        reviews = driver.find_elements_by_class_name("reviewText")
        if len(reviews) >= 3:
            for k in range(3):
                detail_dict['review ' + str(k + 1)] = reviews[k].text
        else:
            k = 0
            while k < len(reviews):
                detail_dict['review ' + str(k + 1)] = reviews[k].text
                k += 1
            while k < 3:
                detail_dict['review ' + str(k + 1)] = None
                k += 1
        all_books_details[book_name] = detail_dict
        driver.back()
```

```python
import pandas as pd
df = pd.DataFrame(all_books_details)
df1 = df.T
df1
```
