```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("/home/rockbot/Documents/Data_Structure_py/mlLab/mlLab code/ConAi/IMDBDataset.csv",nrows=500)
df=df.drop('sentiment',axis=1)
df
```

```python
corpus=[]
for i in range(len(df)):
    corpus.append(df['review'][i])
```

```python
corpus_final=[]
for i in range(len(corpus)):
    corpus_final.append(" ".join([word.lower() for word in corpus[i].split() if word.isalpha()]))
corpus_final
```

```python
document_frequency={}
for i in range(len(corpus_final)):
    word_frequency={}
    for word in corpus_final[i].split():
        word_frequency[word]=1
    document_frequency[i]=word_frequency
```

```python
bow=pd.DataFrame(document_frequency)
bow.fillna(0,inplace=True)
bow
```

```python
cooccurence_matrix=bow.values.dot(bow.values.T)
```

```python
np.fill_diagonal(cooccurence_matrix,0)
```

```python
cooccurence_matrix
```

```python
cooccurence_matrix_t=cooccurence_matrix.T
cooccurence_matrix
```

```python
x=cooccurence_matrix.dot(cooccurence_matrix_t)
x
```

```python
eig_val,eig_vector=np.linalg.eig(x)
```
