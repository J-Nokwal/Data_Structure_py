```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data',na_values='?')
df.columns=["symboling", "normalized_losses","make", "fuel_type", "aspiration","num_doors", "body_style", "drive_wheels","engine_location", "wheel_base", "length", "width", "height", "curb_weight","engine_type", "num_cylinders", "engine_size", "fuel_system", "bore", "stroke","compression_ratio", "horsepower", "peak_rpm", "city_mpg", "highway_mpg", "price"]
df
```

```python
print(df.shape)
df.dropna(subset=['price'],inplace=True)
df.reset_index(drop=True,inplace=True)
print(df.shape)
```

```python
numericalColumns=list(df.dtypes.index[df.dtypes!=object])
df1=df[numericalColumns].fillna(df[numericalColumns].mean())
df1
```

```python
categoricalColumns=list(df.dtypes.index[df.dtypes==object])
df2=df[categoricalColumns].fillna(df[categoricalColumns].mode())
df2
```

```python
categoricalColumns
```

```python
df.update(df1)
df.update(df2)
del df1
del df2
df
```

```python
nancols=[]
for col in df.columns:
    if len(df[col].isnull().unique()) > 1:
        print(col)
        nancols.append(col)
print(nancols)
```

```python
df['num_doors']=df['num_doors'].replace(np.nan,df['num_doors'].mode()[0])
if df.isnull().sum().sum() == 0:
    print('Central Tendency Imputation done')
```

```python
dummies=pd.get_dummies(df[["body_style", "drive_wheels"]])
df=pd.concat((df,dummies),axis=1)
df.drop(columns=["body_style", "drive_wheels"],inplace=True)
df
```

```python
encodedColumns=list(dummies.columns)
categoricalColumns.remove('body_style')
categoricalColumns.remove('drive_wheels')
categoricalColumns,encodedColumns
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
encodingFeatures=['make', 'aspiration', 'engine_location','fuel_type']
for attributes in encodingFeatures:
    df[attributes]=pd.Series(le.fit_transform(df[attributes]))
df
```

```python
categoricalColumns=[x for x in categoricalColumns if not x in encodingFeatures]
encodedColumns+=encodingFeatures
categoricalColumns,encodedColumns
```

```python
def keyEncode(string,substr):
    if string.count(substr):
        return(1)
    else:
        return(0)
```

```python
df[['fuel_system']]=df[['fuel_system']].applymap(lambda x: keyEncode(x,'pfi'))
df[['engine_type']]=df[['engine_type']].applymap(lambda x: keyEncode(x,'ohc'))
categoricalColumns.remove('fuel_system')
categoricalColumns.remove('engine_type')
encodedColumns+=['fuel_system','engine_type']
df
```
