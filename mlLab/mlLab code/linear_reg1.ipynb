<H1> Linear  regression with 1 variable with multiple regression method  </H1>

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([[1,1400],[1,1600],[1,1700],[1,1875],[1,1100],[1,1550],[1,2350],[1,2450],[1,1425],[1,1700]])
y=np.array([[245],[312],[279],[308],[199],[219],[405],[324],[319],[255]])
```

```python
ans=np.linalg.inv(x.T.dot(x)).dot(x.T)
# print(ans.shape)
# print(np.array([[245],[312],[279],[308],[199],[405],[324],[319],[255]]).shape)
b=ans@y
print(b[0][0])
b[0][0]*3000
```

```python

plot_x=np.arange(x.min(),x.max())
plot_y=plot_x*b[0]
plt.plot(plot_x,plot_y)
plt.show()
```
