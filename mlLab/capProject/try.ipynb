```python
import numpy as np
import pandas as pd
import nltk 
from nltk.tokenize import sent_tokenize,word_tokenize
import sklearn
from data1 import topics
from data2 import qes
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from nltk.stem.wordnet import WordNetLemmatizer
import re
```

```python
clean_text_1=[words.lower() for words in topics]
clean_text_2=[sent_tokenize(sent) for sent in clean_text_1 ]
```

```python
clean_text_3 =[]
for word in clean_text_2:
    clean=[]
    for w in word:
        res= re.sub(r'[^\w\s]',"",w)
        if res != "":
            clean.append(res)
    clean_text_3.append(clean)
```

```python
clean_text_4=[]
for words in clean_text_3:
    w=[]
    for word in words:
        if not word in stopwords.words('english'):
            w.append(word)
    clean_text_4.append(w)
```

```python
poter =PorterStemmer()
lem=[]
for words in clean_text_4:
    w=[]
    for word in words:
        w.append(poter.stem(word))
    lem.append(w)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
corpus=topics
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(corpus)
```
