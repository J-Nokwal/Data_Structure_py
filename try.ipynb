```python
from collections import Counter
class Solution:
    def maximumSubsequenceCount(self, text: str, pattern: str) -> int:
        pattern= list(pattern)
        text=list(text)
        count=0
        l=[]
        for i in range(len(text)):
            if text[i] in pattern:
                l.append(text[i])
        # l1=[pattern[0]]+l
        # l2=l+[pattern[1]]
        le=len(l)
        print(le)
        s="".join(pattern)
        mp=Counter(l)
        for i in range(le-1):
            for j in range(i+1,le):
                if l[i]+l[j]==s:
                    count+=1
        # for i in range(le-1):
            # for j in range(i+1,le):
                # if l2[i]+l2[j]==s:
                    # count2+=1
        return max(mp.values())+count
        
a=Solution()
# b="abdcdbc", "ac"
b=""

c=a.maximumSubsequenceCount(b, "uc")
print(c)
```
